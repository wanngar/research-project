```python
import pandas as pd
```

```python
genes = pd.read_csv('RNA-seq of TCGA-LUAD.csv',index_col='Unnamed: 0')

survive_and_time = pd.read_csv("survival of TCGA-LUAD.csv",index_col='Unnamed: 0')
```

```python
genes.head()
```

```python
#status: 0-dead, 1-survived
#os: is the time of life from the beginning of diagnosis until death

survive_and_time.head()
```

```python
genes.index.values
```

```python
arr = ["STAT2", "STAT1", "JAK2", "STAT5A", "JAK1", "SRC", "E2F8", "E2F7", "E2F6", "E2F5", "E2F4", "E2F2", "E2F1", "CDKN2A",
            "CDKN2B", "CDK4", "MYC", "CCND1", "CDKN1B", "CCND2", "CDK6", "CDKN1A", "CDC25A", "CDK2", "CCNE1", "STAT3", "STAT5B",
            "E2F3", "CDK1", "CCNB1", "CCNA1", "RBL2", "RBL1", "RB1" ]

genes = genes[genes.index.isin(arr)]

genes.head()
```

```python
genes = genes.T

genes['type_sample'] = [id.split('-')[-1] for id in genes.index.values]
genes.index = ['-'.join(id.split('-')[:-1]) for id in genes.index.values]

genes.head()
```

```python
data = pd.merge(genes,survive_and_time,how='inner',left_on=genes.index,right_on=survive_and_time.index)
data.rename(columns={"key_0":"id","os":"os_days","status":"survived"},inplace=True)
data.head()
```

```python
data.to_csv('new_genes_data',index=False)
```
