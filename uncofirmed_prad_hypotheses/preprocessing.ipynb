## Считываем данные в датафрейм

```python
prad_clinical = pd.read_csv('https://raw.githubusercontent.com/wanngar/research_project/refs/heads/main/prad_data/prad_clinical.csv', index_col='Unnamed: 0')
prad_gens = pd.read_csv('https://github.com/wanngar/research_project/raw/refs/heads/main/prad_data/prad_rnaseq.csv', index_col='Unnamed: 0')
prad_clinical_ = prad_clinical.copy()
prad_gens_ = prad_gens.copy()
```

```python
prad_gens_ = prad_gens_.T
prad_gens_
```

## Объединяем два датафрейма в один (объединяем гены и рецидив)

```python
prad_clinical_ = prad_clinical_['patient.biochemical_recurrence']
data = pd.merge(prad_gens_,prad_clinical_, how='inner', left_on=prad_gens_.index, right_on=prad_clinical_.index)
```
